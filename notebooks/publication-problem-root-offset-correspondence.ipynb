```python
x=RR['x'].gen()
x.parent()

xsym = var('xsym')
```

```python
gexp=lambda d, i: sum([x**(d*n+i)/factorial(d*n+i) for n in range(60)])
```

```python
def exp_scale(ff, factor):
    def _f(x):
        return ff(x)*exp(factor*x)
    
    return _f
```

```python
d=3

Gd = [gexp(d, (d-j)%d) for j in range(d)]
Gdd = [(lambda x, j=_j: Gd[j](x) - 1/d*exp(x)) for _j in range(d)]

zeta = cos(2*pi/d)+I*sin(2*pi/d)

Jdd = [sum(1/d*exp(zeta^rindex*xsym) for rindex in range(1, d))]
for oindex in range(1, d):
    Jdd.append(Jdd[-1].derivative(xsym))

tbndargs = dict(xmin=-20, xmax=20, ymin=-1.5, ymax=1.5)

plots = [[plot(Gdi, **tbndargs), plot(Jdi, **tbndargs), plot(Jdi(xsym=xsym)*exp(-zeta.real()*xsym), **tbndargs)] for Gdi, Jdi in zip(Gdd, Jdd)]

graphics_array(plots, nrows=d, ncols=3)
```
