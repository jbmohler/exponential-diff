```python
expargs = dict(xmin=-6, xmax=6, ymin=-30, ymax=30)
bndargs = dict(xmin=-16, xmax=16, ymin=-6, ymax=6)
tbndargs = dict(xmin=-1, xmax=32, ymin=-1.5, ymax=1.5)

tbndargs = dict(xmin=-8, xmax=20, ymin=-2.5, ymax=2.5)
```

```python
p = 3
f = [
    sum([x**(p*n+offset)/factorial(p*n+offset) for n in range(40)])
    for offset in range(p)
]

plot(f, **expargs)
```

```python
def hypscale(ff, factor):
    def _f(x):
        return ff(x=x)*(cosh(x*factor))
    
    return _f

print(sum([1 for i in range(p-1) for j in range(i+1, p)]))
labels = [f'$\\mathbb{{f}}^{i}_{p} - \\mathbb{{f}}^{j}_{p}$' for i in range(p-1) for j in range(i+1, p)]
plot([hypscale(f[i]-f[j], 2) for i in range(p-1) for j in range(i+1, p)], **tbndargs, legend_label=labels, figsize=8)
```
