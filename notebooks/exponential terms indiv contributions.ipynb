```python
@interact
def exponential_terms(x=slider(srange(.01,710,.5),default = 50, label="x"), 
                      n = slider([2..80],default = 40, label="# of terms left & right")):
    X = x
    N1 = int(round(X)) - n
    N2 = int(round(X)) + n
    p = scatter_plot([[N, term(X, N).n()] for N in range(max(N1, 0), N2 + 1)])
    show(p)
```
